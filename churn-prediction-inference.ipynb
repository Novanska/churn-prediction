```python
# Load Models
import pandas as pd
import pickle
from tensorflow.keras.models import load_model

with open('final_pipeline.pkl','rb') as file_1:
  model_pipeline = pickle.load(file_1)

model_seq = load_model('model_seq.h5')
model_seq1 = load_model('model_seq1.h5')
model_func = load_model('model_func.h5')
model_func1 = load_model('model_func1.h5')

import warnings
warnings.filterwarnings('ignore')
```

```python
#load data
df = pd.read_csv('D:\Hacktiv8\p2---ftds-019-rmt---m1-Novanska\churn.csv')
```

```python
#mengambil inference dari sample 10 data dari df.
df_inf = df.sample(10)
```

```python
#drop data yang tidak dibutuhkan
df_inf.drop(['churn_risk_score','user_id','internet_option','last_visit_time'],axis=1,inplace=True)
```

```python
df_inf
```

```python
# Transform Inference-Set

df_inf_transform = model_pipeline.transform(df_inf)
df_inf_transform
```

```python
#predict dengan model_sequential
import numpy as np
y_pred_inf = model_seq.predict(df_inf_transform)
y_pred_inf = np.where(y_pred_inf >= 0.5,1,0)
y_pred_inf
```

```python
#predict dengan model_sequential improvement
y_pred_inf = model_seq1.predict(df_inf_transform)
y_pred_inf = np.where(y_pred_inf >= 0.5,1,0)
y_pred_inf
```

```python
#predict dengan model_functional
y_pred_inf = model_func.predict(df_inf_transform)
y_pred_inf = np.where(y_pred_inf >= 0.5,1,0)
y_pred_inf
```

```python
#predict dengan model functional improvement
y_pred_inf = model_func1.predict(df_inf_transform)
y_pred_inf = np.where(y_pred_inf >= 0.5,1,0)
y_pred_inf
```

Insight :

dari data sequential,sequential improvement, terdapat 5 customers yang akan churn, namun menurut functional terdapat 3 customer yang akan churn dan pada functional improvement akan ada 6 customers yang akan churn.
